# **Decada 80 - Question 1**

Which team won the most home games in the 1980s

Qual time ganhou mais jogos em casa na década de 80?

```python
games80_win_home = pd.read_sql(
    """SELECT resultado.team_abbreviation_home, resultado.team_name_home, max(resultado.TOTAL_WIN_HOME) as total_win_home
        FROM
        (SELECT team_abbreviation_home, team_name_home, COUNT(*) AS TOTAL_WIN_HOME
      FROM GAME 
       WHERE GAME.game_date BETWEEN "1980-01-01" AND "1989-12-31"
        AND GAME.wl_home = "W"
 GROUP BY GAME.team_abbreviation_home, GAME.team_id_home) as resultado; """, conn)


print(games80_win_home)
```

**Boston Celtics  - 315 home win**

**Boston Celtics - 315 vitórias em casa**

# **Decada 80 - Question 9**

What was the biggest winner in the clashes of the 80s between the team that won the most games at home and team that won games away from home?


Qual foi o maior vencedor nos confrontos da década de 80 entre o time que mais venceu jogos em casa e time que ganhou jogos fora de casa?

```python
games80_bos_lal = pd.read_sql("""SELECT strftime('%d-%m-%Y',GAME.game_date) as data_game, 
CASE WHEN (GAME.team_abbreviation_home = "BOS" AND GAME.wl_home = "W") THEN "Boston Celtics"            
     WHEN (GAME.team_abbreviation_home = "BOS" AND GAME.wl_home = "L") THEN "Los Angeles Lakers"
     WHEN (GAME.team_abbreviation_home = "LAL" AND GAME.wl_home = "W") THEN "Los Angeles Lakers"
     WHEN (GAME.team_abbreviation_home = "LAL" AND GAME.wl_home = "L") THEN "Boston Celtics"
       END as winner_team,    
GAME.team_abbreviation_home as team_home,GAME.pts_home,
       GAME.team_abbreviation_away as team_away, GAME.pts_away       
 FROM GAME
 WHERE GAME.game_date BETWEEN "1980-01-01" AND "1989-12-31"
  AND (GAME.team_abbreviation_home = "BOS" AND GAME.team_abbreviation_away = "LAL"
   OR GAME.team_abbreviation_home = "LAL" AND GAME.team_abbreviation_away = "BOS"); """, conn)

print(games80_bos_lal)
```
